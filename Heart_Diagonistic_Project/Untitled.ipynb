# THIS ALL WERE CATEGORICAL V/S NUMERICAL GRAPHS

```python
plt.figure(figsize=(4,4))
sns.countplot(df,x='target')
plt.title('Number of females and males')
plt.show()
```

```python
df['sex'].value_counts()
```

```python
female = df['target'][df['sex']==0].value_counts(normalize = True) # filtering female data from target 
feamle_with_disease = female[1]
perc_of_female_with_disease = feamle_with_disease * 100

print(perc_of_female_with_disease)
```

```python
male = df['target'][df['sex']==1].value_counts(normalize = True) # filtering male data from target 
male_with_disease = male[1]
perc_of_male_with_disease = male_with_disease * 100
print(perc_of_male_with_disease)
```

```python
# most of the female have heart disease
 
plt.figure(figsize=(3,3))
sns.barplot(df,x='sex',y='target')
plt.xticks([0,1],['Female','Male'])
plt.xlabel('gender')
print("Percentage of Females with Heart Disease : " , perc_of_female_with_disease)
print("Percentage of males with Heart Disease : " , perc_of_male_with_disease)
plt.title('Comparison between Men and Women Heart Health')
plt.show()
```

```python
# from this graph we can see women have more heart problems where as mens heart is healthy as compared to women
plt.figure(figsize=(5,4))
sns.countplot(df,x='sex',hue='target')
plt.xticks([0,1],['Female','Male'])
plt.legend(labels=['No Disease','Disease'])
plt.title('Gender V/S Healthy Heart V/S Non-Healthy Heart')
plt.show()
```

```python
sns.countplot(df,x='cp')
```

```python
# most of the person who is having 2nd catgory of chest pain i.e Non-angina chest pain are supposed to have heart disease
# while person with 0 category of chest pain does not have much chances of heart disease
plt.figure(figsize=(5,4))
sns.countplot(df,x='cp',hue='target')
#plt.xticks([0,1],['Female','Male'])
plt.legend(labels=['No Disease','Disease'])
plt.title('Having Disease with respect to Chest Pain')
plt.show()
```

```python
# checking which type of chest pain in which Gender ?
# 0 chest pain mostly occur in male as comapred to other type of chest pain this means male have less chance of heat disese

plt.figure(figsize=(5,4))
sns.countplot(df,x='cp',hue='sex')
#plt.xticks([0,1],['Female','Male'])
plt.legend(labels=['female','male'])

plt.title('Chest pain with respect to Gender')
plt.show()
```

```python
plt.figure(figsize=(5,7))
sns.violinplot(df,x='fbs',y = 'age' ,hue='target', split = True)
#plt.xticks([0,1],['Female','Male'])
#plt.legend(labels=['Female','Male'])
plt.title('Fps according to Age')
plt.show()
# females from age 25 to 90 have a chances of having heart disease while most of them have heart disease between age 45 to 60
```
